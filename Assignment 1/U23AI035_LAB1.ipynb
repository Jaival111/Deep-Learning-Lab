## 3. Write a program to implement AND OR gates using Perceptron.

```python
import numpy as np

def step_function(x):
    return 1 if x >= 0 else 0

def perceptron(x, w, b):
    net = np.dot(x, w) + b
    return step_function(net)

inputs = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

# AND gate parameters
w_and = np.array([1, 1])
b_and = -1.5

# OR gate parameters
w_or = np.array([1, 1])
b_or = -0.5

print("AND Gate Output")
for x in inputs:
    print(f"{x} -> {perceptron(x, w_and, b_and)}")

print("\nOR Gate Output")
for x in inputs:
    print(f"{x} -> {perceptron(x, w_or, b_or)}")
```

## 4. Implementation of XOR Problem using PyTorch Neural Network.

```python
import torch.nn as nn
import torch.optim as optim
```

```python
X = torch.tensor([[0., 0.],
                  [0., 1.],
                  [1., 0.],
                  [1., 1.]])

y = torch.tensor([[0.],
                  [1.],
                  [1.],
                  [0.]])
```

```python
class XORNet(nn.Module):
    def __init__(self):
        super(XORNet, self).__init__()
        self.hidden = nn.Linear(2, 2)
        self.output = nn.Linear(2, 1)

    def forward(self, x):
        x = torch.relu(self.hidden(x))
        x = torch.sigmoid(self.output(x))
        return x
```

```python
model = XORNet()

criterion = nn.BCELoss()
optimizer = optim.SGD(model.parameters(), lr=0.1)
```

```python
epochs = 10000

for epoch in range(epochs):
    optimizer.zero_grad()
    
    outputs = model(X)
    loss = criterion(outputs, y)
    
    loss.backward()
    optimizer.step()
    
    if (epoch + 1) % 2000 == 0:
        print(f"Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}")
```

```python
with torch.no_grad():
    predictions = model(X)
    predicted_labels = (predictions > 0.5).float()
    accuracy = (predicted_labels == y).float().mean()

    print("\nFinal Predictions:")
    for i in range(len(X)):
        print(f"Input: {X[i].tolist()}, Predicted: {predictions[i].item():.4f}, Target: {y[i].item()}")

    print(f"\nAccuracy: {accuracy.item()*100:.2f}%")
```

```python
# Input features
X = torch.tensor([[1.0, 2.0],
                  [2.0, 3.0],
                  [3.0, 4.0],
                  [4.0, 5.0]])

# Target output (regression)
y = torch.tensor([[3.0],
                  [5.0],
                  [7.0],
                  [9.0]])
```

```python
class SimpleNN(nn.Module):
    def __init__(self):
        super(SimpleNN, self).__init__()
        self.hidden = nn.Linear(2, 2)   # Hidden layer
        self.output = nn.Linear(2, 1)   # Output layer

    def forward(self, x):
        x = torch.relu(self.hidden(x))  # f(z)
        x = self.output(x)              # Linear output for regression
        return x
```

```python
model = SimpleNN()

criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)
```

```python
epochs = 5000

for epoch in range(epochs):
    optimizer.zero_grad()
    
    predictions = model(X)
    loss = criterion(predictions, y)
    
    loss.backward()
    optimizer.step()
    
    if (epoch + 1) % 1000 == 0:
        print(f"Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.6f}")
```

```python
with torch.no_grad():
    test_input = torch.tensor([[5.0, 6.0]])
    prediction = model(test_input)

print("Predicted Output:", prediction.item())
```
